```python
import pandas as pd
import numpy as np
```

```python
#Read hawaii_measurements.csv
measurements = "Resources/hawaii_measurements.csv"
measurements_df = pd.read_csv(measurements, low_memory=False)

#Print total number of rows
print(len(measurements_df.index))

#Print a groupby to see where data is missing
print(measurements_df.groupby(['station']).count())

measurements_df.head()
```

```python
#Drop all rows with NaN 
clean_measurements_df = measurements_df.dropna(axis=0, how="any")

#Print count of rows dropped
print(f"{len(measurements_df.index) - len(clean_measurements_df.index)} Rows dropped \n")

#Check that there are no more NaNs
print(clean_measurements_df.groupby(['station']).count())

clean_measurements_df.head()
```

```python
#Save clean df as csv
clean_measurements_df.to_csv("Resources/clean_hawaii_measurements.csv", index=False)
```
